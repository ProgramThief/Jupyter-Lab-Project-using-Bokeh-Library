```python
import pandas as pd
import numpy as np  # Import numpy
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
from bokeh.io import push_notebook

# Load the dataset
file_path = 'NBA_Team_Stats.csv'  # Update this path as needed
nba_stats = pd.read_csv(file_path)

# Filter the dataset to include only the specified teams
filtered_teams = nba_stats[nba_stats['Team'].isin(['L.A.Lakers', 'Dallas', 'Golden State'])].copy()

# Convert the "Year" column to numeric by extracting the starting year
filtered_teams.loc[:, 'Year'] = filtered_teams['Year'].apply(lambda x: int(x.split('-')[0]))

# Helper function to calculate histogram data
def get_hist_data(df, col):
    hist, edges = np.histogram(df[col], bins=20, density=True)
    
    # Centering bins
    left = edges[:-1]
    right = edges[1:]
    center = (left + right) / 2
    
    # Create DataFrame
    hist_df = pd.DataFrame({'proportion': hist, 'left': left, 'right': right})
    
    # Create formatted columns
    hist_df['f_interval'] = ['%d - %d' % (left[i], right[i]) for i in range(len(left))]
    hist_df['f_proportion'] = ['%.5f' % prop for prop in hist_df['proportion']]
    
    return hist_df

# Styling function
def style(p):
    # Title
    p.title.align = 'center'
    p.title.text_font_size = '20pt'
    p.title.text_font = 'serif'

    # Axis titles
    p.xaxis.axis_label_text_font_size = '14pt'
    p.xaxis.axis_label_text_font_style = 'bold'
    p.yaxis.axis_label_text_font_size = '14pt'
    p.yaxis.axis_label_text_font_style = 'bold'

    # Tick labels
    p.xaxis.major_label_text_font_size = '12pt'
    p.yaxis.major_label_text_font_size = '12pt'

    return p

# Function to create the plot
def make_plot(src):
    # Blank plot with correct labels
    p = figure(width=700, height=700, 
               title='Histogram of Points per Game by Team',
               x_axis_label='Points per Game', y_axis_label='Count')

    # Quad glyphs to create a histogram
    p.quad(source=src, bottom=0, top='proportion', left='left', right='right',
           color='color', fill_alpha=0.7, hover_fill_color='color', legend_field='name',
           hover_fill_alpha=1.0, line_color='black')

    # Hover tool with vline mode
    hover = HoverTool(tooltips=[('Team', '@name'), 
                                ('Points per Game', '@f_interval'),
                                ('Proportion', '@f_proportion')],
                      mode='vline')

    p.add_tools(hover)

    # Styling
    p = style(p)

    return p

# Prepare the data for plotting
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']
teams = filtered_teams['Team'].unique()
data = pd.DataFrame()

for i, team in enumerate(teams):
    team_data = filtered_teams[filtered_teams['Team'] == team]
    hist_data = get_hist_data(team_data, 'Pts')
    hist_data['name'] = team
    hist_data['color'] = colors[i]
    data = pd.concat([data, hist_data])

# Convert to ColumnDataSource
src = ColumnDataSource(data)

# Create the plot
p = make_plot(src)

# Display the plot in the notebook
output_notebook()
show(p, notebook_handle=True)
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool, CheckboxGroup, CustomJS
from bokeh.layouts import column
from bokeh.io import output_notebook
import pandas as pd
import numpy as np

# Sample data (replace this with your actual dataset)
data = pd.read_csv("NBA Season Data.csv")

# Sample carrier list
carrier_list = ["NOJ", "BUF", "WSB", "BOS", "NYK", "NJN", "CHI", "DEN", "GSW"]

# Prepare data
def make_dataset(carrier_list, range_start=-60, range_end=120, bin_width=5):
    histograms = []
    for carrier in carrier_list:
        subset = data[data['Tm'] == carrier]
        hist, edges = np.histogram(subset['BPM'], bins=int((range_end - range_start) / bin_width), range=(range_start, range_end))
        histograms.append({'carrier': carrier, 'hist': hist, 'left': edges[:-1], 'right': edges[1:]})
    return histograms

# Output to Jupyter Notebook
output_notebook()

# Create a new plot with a title and axis labels
p = figure(title="BPM Histogram", x_axis_label='BPM', y_axis_label='Count')

# Checkbox group widget
checkbox_group = CheckboxGroup(labels=carrier_list, active=list(range(len(carrier_list))))

# Plot each carrier's histogram
colors = ['red', 'green', 'blue', 'orange', 'purple', 'cyan', 'magenta', 'yellow', 'black']
sources = []
for i, carrier in enumerate(carrier_list):
    source = ColumnDataSource(data=dict(hist=[], left=[], right=[]))
    p.quad(source=source, bottom=0, top='hist', left='left', right='right', fill_color=colors[i], line_color='black', legend_label=carrier, alpha=0.7)
    sources.append(source)

# Add hover tool
hover = HoverTool(tooltips=[('BPM', '@left to @right'), ('Count', '@hist'), ('Carrier', '$name')])
p.add_tools(hover)

# Define callback function for checkbox group using JavaScript
checkbox_callback = CustomJS(args=dict(sources=sources), code="""
    const selected_carriers = [];
    for (let i = 0; i < cb_obj.labels.length; i++) {
        if (cb_obj.active.includes(i)) {
            selected_carriers.push(cb_obj.labels[i]);
        }
    }
    for (let i = 0; i < sources.length; i++) {
        if (selected_carriers.includes(sources[i].name)) {
            sources[i].visible = true;
        } else {
            sources[i].visible = false;
        }
    }
""")

checkbox_group.js_on_change('active', checkbox_callback)

# Layout setup
layout = column(p, checkbox_group)

# Show the plot
show(layout)
```

```python
import pandas as pd
import numpy as np
import random
import datetime

# Function to generate random date
def random_date(start_date, end_date):
    return start_date + datetime.timedelta(seconds=random.randint(0, int((end_date - start_date).total_seconds())))

# Generate sample data for temperature, precipitation, and sea level rise variations over time across regions
regions = ['North America', 'Europe', 'Asia', 'Africa', 'South America', 'Australia']

start_date = datetime.datetime(2000, 1, 1)
end_date = datetime.datetime(2023, 12, 31)
date_range = pd.date_range(start=start_date, end=end_date, freq='M')

climate_data = {
    'Date': [],
    'Region': [],
    'Temperature': [],
    'Precipitation': [],
    'Sea_Level_Rise': []
}

for date in date_range:
    for region in regions:
        climate_data['Date'].append(date)
        climate_data['Region'].append(region)
        climate_data['Temperature'].append(random.uniform(-10, 40))  # Temperature in Celsius
        climate_data['Precipitation'].append(random.uniform(0, 500))  # Precipitation in mm
        climate_data['Sea_Level_Rise'].append(random.uniform(0, 20))  # Sea level rise in cm

# Generate sample data for extreme weather events over time
events = ['Hurricane', 'Drought', 'Heatwave', 'Flood']

extreme_events_data = {
    'Date': [],
    'Location': [],
    'Event': [],
    'Severity_Level': []
}

for _ in range(200):
    extreme_events_data['Date'].append(random_date(start_date, end_date))
    extreme_events_data['Location'].append(random.choice(regions))
    extreme_events_data['Event'].append(random.choice(events))
    extreme_events_data['Severity_Level'].append(random.randint(1, 5))  # 1-5 severity level

# Generate sample data for carbon emission footprints
carbon_emission_data = {
    'Year': [],
    'Region': [],
    'Carbon_Emission': []
}

for year in range(2000, 2024):
    for region in regions:
        carbon_emission_data['Year'].append(year)
        carbon_emission_data['Region'].append(region)
        carbon_emission_data['Carbon_Emission'].append(random.uniform(1000, 10000))  # Carbon emission in metric tons

# Create DataFrame
climate_df = pd.DataFrame(climate_data)
extreme_events_df = pd.DataFrame(extreme_events_data)
carbon_emission_df = pd.DataFrame(carbon_emission_data)

# Save to CSV files
climate_df.to_csv('climate_data.csv', index=False)
extreme_events_df.to_csv('extreme_events_data.csv', index=False)
carbon_emission_df.to_csv('carbon_emission_data.csv', index=False)
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
from bokeh.tile_providers import get_provider, Vendors
from bokeh.palettes import Category10
import pandas as pd

# Load the dataset
df = pd.read_csv("climate comp data.csv")

# Remove rows with non-numeric entries in 'Year' column
df = df[df['Year'].str.isnumeric()]

# Convert 'Year' column to datetime format
df['Year'] = pd.to_datetime(df['Year'], format='%Y')

# Create a Bokeh ColumnDataSource
source = ColumnDataSource(df)

# Output to a static HTML file
output_file("carbon_emission_footprints_map.html")

# Define the unique regions and create a color map
regions = df['Region'].unique()
color_map = factor_cmap('Region', palette=Category10[len(regions)], factors=regions)

# Create a figure
p = figure(title="Carbon Emission Footprints Map",
           x_axis_type="mercator", y_axis_type="mercator",
           x_axis_label='Longitude', y_axis_label='Latitude',
           width=800, height=400)

# Add map tile from OpenStreetMap
tile_provider = get_provider(Vendors.CARTODBPOSITRON)
p.add_tile(tile_provider)

# Plot carbon emission footprints as circles on the map with different colors for each region
p.circle(x='Longitude', y='Latitude', source=source, size=10, color=color_map, alpha=0.5, legend_field='Region')

# Add hover tool
hover = HoverTool()
hover.tooltips = [
    ("Year", "@Year{%F}"),
    ("Region", "@Region"),
    ("Carbon Emissions", "@{Carbon Emissions (million metric tons)}{0,0}")
]
hover.formatters = {'@Year': 'datetime'}
p.add_tools(hover)

# Add legend
p.legend.title = 'Region'
p.legend.location = 'top_right'
p.legend.click_policy = "hide"

# Show the plot
show(p)
```
